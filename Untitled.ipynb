```python
import pickle
import librosa
import numpy as np
```

```python
train_spectograms_X = []
train_spectograms_X_complex = []
train_spectograms_S = []
train_spectograms_S_complex = []
train_spectograms_N = []
train_spectograms_N_complex = []

for suffix in range(1200):
  
  print(suffix+1, 'of 1200')
  
  name_x = 'trx'+str(suffix).zfill(4)+'.wav'
  name_s = 'trs'+str(suffix).zfill(4)+'.wav'
  name_n = 'trn'+str(suffix).zfill(4)+'.wav'
  
  s, sr = librosa.load("RNNDenoiseData/data/tr/"+name_x, sr = None)
  x_complex = librosa.stft(s, n_fft = 1024, hop_length=512)
  train_spectograms_X.append(np.abs(x_complex))
  train_spectograms_X_complex.append(x_complex)
  
  s, sr = librosa.load('RNNDenoiseData/data/tr/'+name_s, sr = None)
  s_complex = librosa.stft(s, n_fft = 1024, hop_length=512)
  train_spectograms_S.append(np.abs(s_complex))
  train_spectograms_S_complex.append(s_complex)
  
  s, sr = librosa.load('RNNDenoiseData/data/tr/'+name_n, sr = None)
  n_complex = librosa.stft(s, n_fft = 1024, hop_length=512)
  train_spectograms_N.append(np.abs(n_complex))
  train_spectograms_N_complex.append(n_complex)

train_spectograms_M = []
for s, n, i in zip(train_spectograms_S, train_spectograms_N, range(len(train_spectograms_S))):
  m = np.greater(s,n)
  train_spectograms_M.append(m.astype(int))
```

```python
validation_spectograms_X = []
validation_spectograms_X_complex = []
validation_spectograms_S = []
validation_spectograms_S_complex = []
validation_spectograms_N = []
validation_spectograms_N_complex = []

for suffix in range(1200):
  
  print(suffix+1, 'of 1200')
  
  name_x = 'vx'+str(suffix).zfill(4)+'.wav'
  name_s = 'vs'+str(suffix).zfill(4)+'.wav'
  name_n = 'vn'+str(suffix).zfill(4)+'.wav'
  
  s, sr = librosa.load("RNNDenoiseData/data/v/"+name_x, sr = None)
  x_complex = librosa.stft(s, n_fft = 1024, hop_length=512)
  validation_spectograms_X.append(np.abs(x_complex))
  validation_spectograms_X_complex.append(x_complex)
  
  s, sr = librosa.load('RNNDenoiseData/data/v/'+name_s, sr = None)
  s_complex = librosa.stft(s, n_fft = 1024, hop_length=512)
  validation_spectograms_S.append(np.abs(s_complex))
  validation_spectograms_S_complex.append(s_complex)
  
  s, sr = librosa.load('RNNDenoiseData/data/v/'+name_n, sr = None)
  n_complex = librosa.stft(s, n_fft = 1024, hop_length=512)
  validation_spectograms_N.append(np.abs(n_complex))
  validation_spectograms_N_complex.append(n_complex)

validation_spectograms_M = []
for s, n, i in zip(validation_spectograms_S, validation_spectograms_N, range(len(validation_spectograms_S))):
  m = np.greater(s,n)
  validation_spectograms_M.append(m.astype(int))
```

```python
test_spectograms_X = []
test_spectograms_X_complex = []


for suffix in range(1200):
  
  print(suffix+1, 'of 1200')
  
  name_x = 'tex'+str(suffix).zfill(4)+'.wav'

  s, sr = librosa.load("RNNDenoiseData/data/te/"+name_x, sr = None)
  x_complex = librosa.stft(s, n_fft = 1024, hop_length=512)
  test_spectograms_X.append(np.abs(x_complex))
  test_spectograms_X_complex.append(x_complex)
  
```
